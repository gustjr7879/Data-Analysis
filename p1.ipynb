- 데이터 프레임은 행과 열로 구성되어있는 데이터이다. 과연 데이터가 많을때(행 또는 열이 많다.) 행이 많은게 좋을까, 열이 많은게 좋을까?    
- 정답은 열이 많은 것이 더 좋다고 한다. 행이 많아지더라도 분석 기술 면에서는 별다른 차이가 생기지 않는다.   
- 하지만 열이 많아진다면 분석할 때 변수를 조합할 수 있는 경우의 수가 늘어난다.   
- 예를 들어서 학점 별 연봉 관계 를 분석하다가 열로 지역이 들어왔다고 해보자.    
- 이럴 때 더 다채롭게 조합해서 분석할 수 있다.   

```python
import pandas as pd

df = pd.DataFrame({'학번': [1,2,3],
                   'eng score': [90,80,60],
                   'math score': [40,50,80]})
print(df)
```

```python
#sum을 이용해서 english score 합계 계산
print(df['eng score'].sum())
#describe를 사용해서 여러가지 통계 확인
print(df.describe())
```
