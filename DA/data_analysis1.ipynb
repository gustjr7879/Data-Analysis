```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

```python
sns.set(color_codes = True)
%matplotlib inline
```

```python
df = pd.read_csv('/Users/hyunskki/codezip/DA/hotel_bookings.csv')
df.head()
```

```python
# 각 컬럼의 속성 및 결측치 확인
# info 는 데이터의 전반적인 정보를 나타낸디. 몇개가 nan 데이터 인지도 보여준다.  
df.info()
```

```python
#describe는 통계치를 보여준다. 
df.describe()
```

```python
#skew 는 왜도를 보여준다 (왜도란 좌우로 치우쳐져있는지 보여줌)
#수치형만 보여줄 수 있기 때문에 몇개만 골라서 예시로 보여줌
#0보다 작거나 3보다 크면 데이터가 정규성을 가지지 않는다고 판단한다.
df[['is_canceled','lead_time','adults']].skew()
#변숫값 분포의 정규성이 필요한 경우 로그변환, 정규화 표준화 등을 진행해본다. 
```

```python
#kurtosis 는 첨도를 확인해준다.
df[['is_canceled','lead_time','adults']].kurtosis()
```

```python
#시각화를 통해 특정 변수의 분포를 확인
sns.distplot(df['lead_time'])
#이 결과를 통해서 0값이 많은 것을 확인했고, 당일 체크인 하는 투숙객이 많다는 것을 확인할 수 있다. 
```

```python
#그룹 구분에 따른 특정 변수 분포 차이 시각화
sns.violinplot(x='hotel',y='lead_time',data=df,inner=None,color=".8")
sns.stripplot(x='hotel',y='lead_time',data=df,size=1)
#시각화는 다양한 방식으로 구현할 수 있다. 
#호텔 구분에 따라 투숙객의 lead_time이 어떻게 되는지 시각화할 수 있다. violinplot은 분포를 효과적으로 보여준다.
#stripplot는 각 관측치의 위치를 직관적으로 표현해준다. 이 두가지를 동시에 사용하면 데이터의 형태를 보다 잘 이해할 수 있다. 
```

```python
#데이터 탐색 중 특성을 파악한 뒤 항상 진행하는 과정이 변수간의 관계를 파악하는 것이다. 
#타겟 변수와 입력변수 뿐만 아니라 입력 변수간의 관계도 살펴본다. 
#이를 통해서 독립변수의 변화에 따른 종속 변수의 변화량을 크게해서 통계적 정확도를 감소시키는 다중공선성을 방지할 수 있다. 
#상관관계를 파악하는 대표적인 개념으로 공분산과 상관계수가 존재한다. 
#상관분석을 하기 위해서는 데이터가 등간이나 비율 척도이며 두 변수가 선형적관계라는 기본 가정을 둔다. 
#공분산과 상관계수는 얼마나 변동이 닮았는지를 보여주는 것은 동일하지만 계산 과정에서 묘하게 차이가 존재한다. 
#상관 분석은 두 변수간의 선형 관계만 측정할 수 있기 때문에 비선형 관계의 경우 포착하지 못하기 때문에, 산점도를 그려서 함께 확인하는 방법도 사용해야한다.
```

```python
wine_df = pd.read_csv('/Users/hyunskki/codezip/DA/winequality-red.csv')
wine_df.head()
```

```python
#산점도 행렬 시각화
sns.set(font_scale=1.1)
sns.set_style('ticks') # 축 눈금 설정
sns.pairplot(wine_df,
diag_kind='kde') #상관계수가 1이면 분포로 표시
plt.show()
```

```python
#공분산 산출
wine_df.cov()
```

```python
#숫자로 상관계수 표시
wine_df.corr()
```

```python
#heatmap 시각화
sns.heatmap(wine_df.corr(),cmap='viridis')
```

```python
sns.clustermap(wine_df.corr(),annot=True,cmap='RdYlBu_r',vmin=-1,vmax=1)
```

```python
#중복제거 히트맥 생성
import numpy as np
#matrix 우측 상단을 모두 True 하단을 모두 False로
np.triu(np.ones_like(wine_df.corr()))
mask = np.triu(np.ones_like(wine_df.corr(),dtype=bool))

fig,ax = plt.subplots(figsize=(15,10))
sns.heatmap(wine_df.corr(),mask=mask,vmin=-1,vmax=1,annot=True,cbar=True)
ax.set_title('Wine Quality Correlation',pad=15)
```
