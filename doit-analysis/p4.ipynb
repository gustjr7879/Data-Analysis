### 순서대로 정렬하기
데이터를 순서대로 정렬하면 매우 크거나 작은 데이터를 알아낼 수 있다.     
```python
df.sort_values()
```     
를 이용한다

````python
import pandas as pd

exam = pd.read_csv('./exam.csv')
````

````python
exam.sort_values('math') #math 기준으로 오름차순 정렬
````

````python
exam.sort_values('math',ascending=False) #내림 차순 정렬
````

````python
#여러 정렬 기준 적용하기
#정렬 기준으로 삼을 변수를 여러개 지정하려면 []를 사용해서 변수명을 나열하면 된다.
exam.sort_values(['nclass','math']) #nclass를 기준으로 먼저 정렬하고, math를 기준으로 정렬한다
````

````python
#여러 변수를 기준으로 하되 오름차순 내림차순 다르게하기
exam.sort_values(['nclass','math'],ascending=[True,False])
````

````python
#실습 mpg데이터에서 audi에서 생산한 자동차 중에 hwy가 높은 순서대로 출력하기
mpg = pd.read_csv('./mpg.csv')
mpg.sort_values('hwy',ascending=False).query('manufacturer == "audi"')
````

### 파생변수 추가하기
데이터에 들어있는 변수만 이용해서 분석할 수 있지만, 변수를 조합하거나 함수를 이용해서 새 변수를 만들어서 분석할 수 있다.    
df.assign()을 이용하면 데이터에 파생변수를 만들어서 추가할 수 있다.

````python
#exam에 세 과목의 점수를 합한 총점 변수를 추가해 출력한다. exam.assign()에 새 변수명 = 변수를 만드는 공식 을 입력하면 된다.
exam.assign(total = exam['math']+exam['english']+exam['science'])
````

````python
#여러 파생변수를 한번에 추가하기
exam.assign(
    total = exam['math'] + exam['english'] + exam['science'],
    mean = (exam['math'] + exam['english'] + exam['science'])/3
)
````

````python
#df.assign에 np.where 추가하기
#np.where을 추가하면 조건에 따라 다른 값을 부여한 변수를 추가할 수 있다.
import numpy as np
exam.assign(test = np.where(exam['science']>=60,'pass','fail'))
````

````python
#추가한 변수를 pandas 함수에 바로 활용하기
#변수를 추가하고 나면 이어지는 pandas 함수에 바로 활용할 수 있다. 
exam.assign(total = exam['math']+exam['english']+exam['science']).sort_values('total',ascending=False)
````

````python
#lambda를 이용해서 데이터 프레임명을 줄여서 쓰기
long_name = exam.copy()

long_name.assign(new = lambda x : x['english']+x['math']) #long_name 대신 x로 쓸 수 있다.
````

````python
#앞에서 만든 변수 바로 이용해보기
exam.assign(total = exam['math']+exam['english']+exam['science'],
            mean = lambda x : x['total']/3) 
#바로 이용하려면 무조건 lambda를 활용해야한다. 안그러면 오류뜬다 !
````

````python
#실습 
#mpg 데이터 복사본을 만들고 cty와 hwy를 합한 합산 연비 변수를 추가하라
mpg.assign(total = mpg['cty']+mpg['hwy'])
````

````python
#앞에서 구한 합산연비를 2로 나누어서 평균 연비 변수를 추가해라
mpg.assign(total = mpg['cty']+mpg['hwy'],
           mean = lambda x : x['total']/2)
````

````python
#평균 연비가 가장 높은 자동차 3개 출력해라
mpg.assign(total = mpg['cty']+mpg['hwy'],
           mean = lambda x : x['total']/2).sort_values('mean',ascending=False)[:3]
````

### 집단별로 요약하기
집단별 평균이나 집단별 빈도처럼 각 집단을 요약한 값을 구할 때는    
```python
df.groupby()
df.agg()
```     
를 사용한다.

````python
#전체 요약 통계량 구하기
#요약 통계량을 구할 때 df.agg()를 사용한다. df.agg()에 mean_math = ('math','mean') 처럼 요약값을 할당한다.
exam.agg(mean_math = ('math','mean'))
````

````python
#집단별 요약 통계량 구하기
exam.groupby('nclass').agg(mean_math = ('math','mean')) #각 반마다 math 평균값
````

````python
#위의 코드는 nclass가 인덱스가 되어있다. 이를 방지할 수 있다.
exam.groupby('nclass',as_index=False).agg(mean_math = ('math','mean'))
````

````python
#여러 요약 통계량 한번에 구하기
exam.groupby('nclass').agg(mean_math = ('math','mean'),
                           sum_math = ('math','sum'),
                           median_math = ('math','median'))
````

````python
#이를 간단하게 할 수 있는 마성의 코드
exam.groupby('nclass').mean()
````

````python
#groupby()에 여러 변수를 지정하면 집단을 나누고 다시 하위 집단으로 나눌 수 있다.
mpg.groupby(['manufacturer','drv']).agg(mean_cty = ('cty','mean'))
````

````python
#value_counts로 집단별 빈도 간단하게 구하기
mpg.groupby('drv').agg(count = ('drv','count'))
````

````python
mpg['drv'].value_counts() #로 더 간단하게 구할 수 있다.
````

````python
#실습
#제조회사별로 suv 자동차의 도시 및 고속도로 합산 연비 평균을 구해 내림차순으로 정렬하고 1-5위 출력하기
mpg.query('category == "suv"').assign(mean = (mpg['cty']+mpg['hwy'])/2).groupby('manufacturer').agg(mean_tot = ('mean','mean')).sort_values('mean_tot',ascending=False)
````

````python
#실습
#mpg데이터의 category 중 어떤 차종이 cty가 높은지 알아보자
mpg.groupby('category').agg(mean = ('cty','mean')) #subcompact가 가장 높다
````

````python
#어떤 차종이 가장 높은지 알기 편하게 정렬
mpg.groupby('category').agg(mean = ('cty','mean')).sort_values('mean',ascending=False)
````

````python
#어떤 회사 자동차 hwy가 가장 높을까? 
mpg.groupby('manufacturer').agg(mean = ('hwy','mean')).sort_values('mean',ascending=False).head(3)
````

````python
#어떤 회사에서 compact 차종을 가장 많이 생산할까?
mpg.query('category == "compact"').groupby('manufacturer').agg(count = ('category','count'))
````

### 데이터 합치기 
하나의 데이터만 가지고 분석하기도 하지만 여러 데이터를 합쳐서 하나로 만든 다음에 분석하기도 한다.     
예를 들어서 중간고사 데이터와 기말고사 데이터를 합쳐서 하나로 만들면 기말고사 점수가 중간고사 점수보다 올랐는지 알 수 있다.     

````python
#가로로 합치기
test1 = pd.DataFrame({'id':[1,2,3,4,5],'midterm':[60,80,70,90,85]})
test2 = pd.DataFrame({'id':[1,2,3,4,5],'final':[70,80,65,95,80]})

#데이터를 가로로 합칠때는 merge를 사용한다.
#오른쪽에 입력한 데이터 프레임을 왼쪽 데이터 프레임에 결합하도록 하면 how='left'로 해주면 된다. 데이터를 합칠 때 기준으로 삼을 변수명을 on에 입력한다.

total = pd.merge(test1,test2,how='left',on='id')
total
````

````python
#다른 데이터를 활용해 변수 추가하기
#pd.merge()를 이용하면 특정 변수의 값을 기준으로 다른 데이터를 추가할 수 있다. 
name = pd.DataFrame({'nclass':[1,2,3,4,5],'teacher':['k','l','p','c','j']})
#nclass 기준으로 name의 teacher 변수를 exam에 넣어보자
exam_new = pd.merge(exam,name,how='left',on='nclass')
print(exam_new)
````

````python
#세로로 합치기
#학생 5명이 시험을 보고 조금 뒤에 다른 5명이 시험을 봤다고 해보자
#이런 경우에는 세로로 합쳐줘야한다.
g_a = pd.DataFrame({'id':[1,2,3,4,5],'test':[10,20,30,40,50]})
g_b = pd.DataFrame({'id':[6,7,8,9,10],'test':[20,30,40,50,60]})

#이런 경우에는 pd.concat()을 이용한다.
total_g = pd.concat([g_a,g_b],ignore_index=True) #ignore_index를 통해서 인덱스 초기화가 가능하다.
print(total_g)
#pd.concat을 이용해서 데이터를 세로로 합칠때는 두 데이터의 변수 명이 같아야한다. 만약 변수명이 다르면 pd.rename()을 사용해서 맞춰줄 수 있다.
````

````python
#실습
#mpg데이터에 fl는 자동차에 사용하는 연료를 의미한다.
fuel = pd.DataFrame({'fl':['c','d','e','p','r'],'price_fl':[2.35,2.38,2.11,2.76,2.22]})

#fl에 해당하는 가격을 넣어라
pd.merge(mpg,fuel,how='left',on='fl')
````
